```
import pandas as pd
df=pd.read_csv('/content/homeprices.csv')
df
```

```
ndf=pd.get_dummies(df[['town']])
ndf
```

```
finaldf=df.join(ndf)
finaldf
```

```
df=finaldf.drop(['town','town_west windsor'],axis='columns')
df
```

```
x=df.drop(['price'],axis='columns')
x
```

```
y=df[['price']]
y
```

```
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
```

```
lreg.fit(x,y)
lreg
```

```
y_pred=lreg.predict(x)
```

```
lreg.score(x,y)
```
