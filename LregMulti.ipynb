```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```
dic={'area':[2600,3000,3200,3600,4000],
     'bedrooms':[3,4,None,3,5],
'age':[20,15,18,30,8],
'price':[550000,565000,610000,595000,760000]}
df=pd.DataFrame(dic)
df
```

```
a=df['bedrooms']
a=a.median()
a=int(a)
a
```

```
df=df.fillna(a)
df
```

```
lreg=LinearRegression()
lreg.fit(df[['area','bedrooms','age']],df[['price']])
y_pred=lreg.predict(df[['area','bedrooms','age']])
print(y_pred)                                     #  dependent  =   m*independents +   intercept
print(lreg.coef_)                                 #  y=  m1*x1  +  m2*x2  +  m3*x3   + C
print(lreg.intercept_)                            #  526000 =137.25*(2600)+(-26025.)*(3.0)+(-6825.)*20 + 383725.
```

```
df=pd.read_csv('/content/hiring.csv')
exp=df[['experience']].fillna('zero')
df=df.iloc[:,1:]
df=exp.join(df)
test='test_score(out of 10)'
df[test]=df[test].fillna(df[test].median())
df
```

```
li=[0,0,5,2,7,3,10,11]
newdf=pd.DataFrame(li,columns=['experience'])
df=newdf.join(df.iloc[:,1:])
df
```

```
lreg2=LinearRegression()
x=df.iloc[:,:3]
y=df.iloc[:,3:]
lreg2.fit(x,y)
lreg2.predict(x)
```

```
lreg2.predict([[2,9,6]])
```

```
lreg2.predict([[12,10,10]])
```
