```
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```
flow=load_iris()
```

```
dir(flow)
```

```
x=flow['data']
```

```
y=flow['target']
```

```
print(cross_val_score(LogisticRegression(),x,y))
print(cross_val_score(SVC(),x,y))
print(cross_val_score(RandomForestClassifier(n_estimators=500),x,y))
```

we get to know by this for this particular model logistic regression is better than svm is better than random forest with 500 trees
