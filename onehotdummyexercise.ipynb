```
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```
df=pd.read_csv('/content/drive/My Drive/carprice.csv')
df
```

```
ndf=pd.get_dummies(df[['Car Model']])
ndf
```

```
ndf=ndf.iloc[:,1:]
ndf
```

```
df=df.join(ndf)
```

```
df=df.iloc[:,1:]
```

```
df
```

```
x=df.drop(df[['Sell Price($)']],axis='columns')
x
```

```
y=df[['Sell Price($)']]
y
```

```
lreg=LinearRegression()
lreg.fit(x,y)
```

```
lreg.predict([[45000,4,0,1]])
```

```
lreg.predict([[86000,7,1,0]])
```

```
lreg.score(x,y)
```
